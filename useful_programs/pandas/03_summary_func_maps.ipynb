```python
means = file["WaterTemp (C)"].mean()
```

```python
means
```

```python
file["WaterTemp (C)"].map(lambda p: p - means) # Series
# lamda p --> data points 
```

```python
def remean_points(row):
   row["WaterTemp (C)"] = row["WaterTemp (C)"] - means
   return row
```

```python
file.apply(remean_points, axis="columns")
```
